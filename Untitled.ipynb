# Randomness
In most cases, the random varaibles we generate from computers are pseudo-random, the real randomness comes from the nature. In this notebook, we will go through:
* How to transform binary random variable to uniform variable
* How to test the randomness of a random variable generater 

```python
import numpy as np
import matplotlib.pyplot as plt
Sample = np.random.uniform(0,1, 1000)
plt.hist(Sample,facecolor = "white")
plt.show()
```
